```python
import pandas as pd
import random
from faker import Faker
from datetime import datetime, timedelta

fake = Faker()

# Number of records
NUM_RECORDS = 30000

industries = ['IT', 'Finance', 'Healthcare', 'Education', 'Retail']
regions = ['North', 'South', 'East', 'West']
traffic_sources = ['Email', 'Ads', 'Webinar', 'Organic']
company_sizes = ['Small', 'Medium', 'Large']

data = []

for _ in range(NUM_RECORDS):
    
    # Basic Info
    name = fake.name()
    company = fake.company()
    industry = random.choice(industries)
    region = random.choice(regions)
    company_size = random.choice(company_sizes)
    
    # Behavioral Data
    traffic_source = random.choice(traffic_sources)
    site_visits = random.randint(1, 50)
    session_duration = round(random.uniform(1, 30), 2)  # minutes
    
    # Interaction Data
    emails_opened = random.randint(0, 20)
    demo_requested = random.choice([0, 1])
    
    # Time Data
    first_interaction = fake.date_between(start_date='-1y', end_date='today')
    days_since_first = (datetime.now().date() - first_interaction).days
    
    # Target Label (simple logic)
    if (site_visits > 10 and emails_opened > 5 and demo_requested == 1):
        converted = 1
    else:
        converted = random.choice([0, 1])
    
    data.append([
        name, company, industry, region, company_size,
        traffic_source, site_visits, session_duration,
        emails_opened, demo_requested,
        days_since_first, converted
    ])

# Create DataFrame
columns = [
    'Name', 'Company', 'Industry', 'Region', 'Company Size',
    'Traffic Source', 'Site Visits', 'Session Duration',
    'Emails Opened', 'Demo Requested',
    'Days Since First Interaction', 'Converted'
]

df = pd.DataFrame(data, columns=columns)

# Save to CSV
df.to_csv("lead_data.csv", index=False)
print("Dataset Created Successfully ✅")
print(df.head())
```
